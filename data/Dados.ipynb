```python
import wbgapi as wb
import pandas as pd

# 1. Definir os países (PALOP)
palop = ['AGO', 'CPV', 'GNB', 'MOZ', 'STP', 'GNQ']

# 2. Definir o indicador do PIB per capita
indicator = 'NY.GDP.PCAP.CD'

# 3. Consumir a API definindo o intervalo de anos (range)
# O parâmetro 'time' aceita um range de anos
df_pib_longo = wb.data.DataFrame(indicator, palop, time=range(2010, 2026))

# Limpar o DataFrame
df_pib_longo = df_pib_longo.reset_index()

# Visualizar o resultado
print(df_pib_longo)
```

População

```python
import wbgapi as wb

# 1. Definir os países e o indicador de população
paises = ['AGO', 'CPV', 'GNB', 'MOZ', 'STP', 'GNQ']
pop_indicator = 'SP.POP.TOTL'

# 2. Baixar os dados (de 2010 a 2025 para bater com o seu IDH)
df_pop = wb.data.DataFrame(pop_indicator, paises, time=range(2010, 2026))

# 3. Organizar o DataFrame da população (formato longo)
df_pop = df_pop.stack().reset_index()
df_pop.columns = ['iso_code', 'ano', 'populacao_total']

# 4. Limpar o campo 'ano' (remover o 'YR' que a API coloca)
df_pop['ano'] = df_pop['ano'].str.replace('YR', '').astype(int)

display(df_pop.head())
```

```python
import pandas as pd
import json

path = "/content/drive/MyDrive/TCC/dados_HDI/hdr-data.json"

with open(path, 'r', encoding='utf-8') as f:
    dados_json = json.load(f)

df = pd.DataFrame(dados_json)

df_limpo = df[['countryIsoCode', 'country', 'indicatorCode', 'indicator', 'year', 'actualValue']].copy()
df_limpo.columns = ['iso_code', 'pais', 'id_indicador', 'nome_indicador', 'ano', 'valor_real']

path_saida = "/content/drive/MyDrive/TCC/dados_HDI/dados_hdi_palop.csv"
df_limpo.to_csv(path_saida, index=False, encoding='utf-8-sig')

print(f"Sucesso! O arquivo foi transformado e salvo em: {path_saida}")
```

```python
import pandas as pd

df_hdi = df.dropna(subset=['countryIsoCode'])

df_hdi['actualValue'] = pd.to_numeric(df_hdi['actualValue'], errors='coerce')

df_features = df_hdi.pivot_table(
    index=['countryIsoCode', 'country', 'year'],
    columns='indicatorCode',
    values='actualValue'
).reset_index()

df_features.columns = ['iso_code', 'pais', 'ano', 'renda_nacional', 'expectativa_vida']

display(df_features.head())
```
