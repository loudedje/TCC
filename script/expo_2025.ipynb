```python
import pandas as pd
```

O_ANO: Ano de referência em que a operação de exportação foi registrada.

CO_MES: Mês de referência da operação (numérico, de 01 a 12).

SH4: Código de 4 dígitos do Sistema Harmonizado (representa a categoria ou grupo do produto).

CO_PAIS: Código numérico que identifica o país de destino da mercadoria.

SG_UF_MUN: Sigla da Unidade Federativa (Estado) onde está localizada a empresa que realizou a declaração.

CO_MUN: Código numérico do município (padrão IBGE) da empresa declarante.

KG_LIQUIDO: Peso líquido total da mercadoria em quilogramas (volume físico).

VL_FOB: Valor total da mercadoria em Dólares Americanos (US$) no incoterm FOB (Free on Board).
