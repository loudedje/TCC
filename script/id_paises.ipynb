```python
import pandas as pd
import json

with open('dados.json', 'r', encoding='utf-8') as f:
    dados_brutos = json.load(f)

df = pd.json_normalize(dados_brutos)

df = df.replace(r'\r', '', regex=True)

# 4. Salva em CSV
df.to_csv('saida_do_postman.csv', index=False, sep=';', encoding='utf-8-sig')

print("Ficheiro CSV gerado com sucesso!")
```
